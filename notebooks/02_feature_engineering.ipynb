```python
import pandas as pd

df_hourly = pd.read_csv(
    "hourly_power_consumption.csv",
    parse_dates=["datetime"],
    index_col="datetime"
)

df_hourly.head()
```

```python
df_feat = df_hourly.copy()

df_feat["lag_1"] = df_feat["Global_active_power"].shift(1)
df_feat["lag_24"] = df_feat["Global_active_power"].shift(24)
df_feat["lag_168"] = df_feat["Global_active_power"].shift(168)  # 7 days
```

```python
df_feat["roll_mean_24"] = (
    df_feat["Global_active_power"]
    .rolling(window=24)
    .mean()
)

df_feat["roll_mean_168"] = (
    df_feat["Global_active_power"]
    .rolling(window=168)
    .mean()
)
```

```python
df_feat["hour"] = df_feat.index.hour
df_feat["day_of_week"] = df_feat.index.dayofweek
df_feat["is_weekend"] = df_feat["day_of_week"].isin([5, 6]).astype(int)
```

```python
df_feat.dropna(inplace=True)
df_feat.head()
```

```python
df_feat.to_csv("hourly_features.csv")
```
